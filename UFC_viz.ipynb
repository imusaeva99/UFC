```python
with open('events.html', 'r') as f:
  html_all_events = f.read()
```

```python
soup = BeautifulSoup(html_all_events)
```

```python
all_links_to_events = soup(class_ = "b-link b-link_style_black")
```

```python
links = []

for i in all_links_to_events:
  links.append(i['href'])
```

```python
len(links)
```

```python
def get_csv(link):

  url = link
  html = requests.get(url).content
  df_list = pd.read_html(html)
  df = df_list[-1]
  soup = BeautifulSoup(html)
  title_list = soup(class_ = "b-content__title-highlight")
  for t in title_list:
    title = t.text
  df['Event'] = title

  return df
```

```python
url = links[0]
html = requests.get(url).content
df_list = pd.read_html(html)
df_base = df_list[-1]
soup = BeautifulSoup(html)
title_list = soup(class_ = "b-content__title-highlight")
for t in title_list:
  title = t.text
df_base['Event'] = title
```

```python
df_base
```

```python
df = get_csv(links[1])
df_base.append(df, ignore_index=True)
```

```python
df_base
```

```python
from tqdm import tqdm
```

```python
dfs = []

for link in tqdm(links[1:]):
  df = get_csv(link)
  dfs.append(df)

df_big = df_base.append(dfs, ignore_index='True')
```

```python
df_big
```

```python
df_big.to_csv('all_fights.csv')
```
